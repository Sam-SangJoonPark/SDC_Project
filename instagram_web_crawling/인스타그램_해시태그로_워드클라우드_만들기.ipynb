```python
a = pd.read_csv("C:/Users/ASIA/Downloads/like_tag_힙지로.csv")
b = pd.read_csv("C:/Users/ASIA/Downloads/cafe_like_tag.csv")
c = pd.read_csv("like_tag_을지로맛집.csv")
```

#### 각 파일에서 '0' 칼럼은 '좋아요 수'를 뜻함

```python
# a는 이런 데이터프레임
a
```

```python
# a 를 전처리하여 태그 하나씩을 다 리스트 안에 넣자 !
def process(a) :
    for i in range(len(a)):
        a['1'][i] = str(a['1'][i]).replace('[','').replace(']','').replace("'","").replace(',',' ').replace('#', '')
    bbb = []
    for i in range(len(a)) :
        b = a['1'][i].split()
        
        for k in b :
            z = str(k.split())
            bbb.append(z)
    for i in bbb :
        i = i.replace('[','').replace(']','').replace("'","").replace(',',' ')
    lll = []
    for i in bbb :
            i = i.replace('[','').replace(']','').replace("'","").replace(',',' ')
            lll.append(i)
    return lll
```

```python
# 각각의 csv파일들을 같은 방식으로 하자!
a  = process(a)
b= process(b)
c = process(c)
```

```python
# 이런식으로 태그가 나옴
a
```

```python
final = a+b+c
```

```python
# 최종 파일은 이렇게 합쳐짐
len(final)
final
```

```python
# 리스트에 있는 것을 딕셔너리로!!
from collections import Counter 
count = Counter(final) 
common_tag_200 = count.most_common(100)
common_tag_200
```

```python
# 이를 워드클라우드로 표현해보자!
from PIL import Image 

import matplotlib.pyplot as plt 
from wordcloud import WordCloud 
import numpy as np 
from PIL import Image 
import matplotlib 
matplotlib.rcParams['font.family'] = "Maulgun Gothic" 


font_path= 'C:\\WINDOWS\\Fonts\\\x7f\x7f\x7f\x7fEXTRABOLD.TTF'


denne_mask = np.array(Image.open('C:/Users/ASIA/Downloads/alice_mask.png')) 

wc = WordCloud(font_path=font_path, background_color="white", width=800, height=600, mask = denne_mask) 
cloud = wc.generate_from_frequencies(dict(common_tag_200)) 
plt.figure(figsize = (20, 16)) 
plt.axis('off') 
plt.imshow(cloud)
```
